```python
########### Data Preprocessing ############
```

```python
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data["Division Name"].value_counts()
```

```python
data["Division Name"].fillna('General',inplace=True)
```

```python
data["Department Name"].value_counts()
```

```python
data["Department Name"].fillna('Tops',inplace=True)
```

```python
data["Class Name"].value_counts()
```

```python
data["Class Name"].fillna('Dresses',inplace=True)
```

```python
labelDN = LabelEncoder()
data["Division Name"] = labelDN.fit_transform(data["Division Name"].astype(str))

labelDeptN = LabelEncoder()
data["Department Name"] = labelDeptN.fit_transform(data["Department Name"].astype(str))

labelCN = LabelEncoder()
data["Class Name"] = labelCN.fit_transform(data["Class Name"].astype(str))
```

```python
sc = StandardScaler()
data[["Age","Division Name","Department Name","Class Name"]]=sc.fit_transform(data[["Age","Division Name","Department Name","Class Name"]])
```
